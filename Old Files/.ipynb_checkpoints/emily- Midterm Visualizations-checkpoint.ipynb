### Correlation analysis - autocorrelation and partial autocorrelation

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

data['date.x'] = pd.to_datetime(data['date.x'])
data.set_index('date.x', inplace=True)

# Extract the 'covid_19_deaths' column
target_variable = 'covid_19_deaths'
ts = data[target_variable]
```

```python
# Plot ACF
plot_acf(ts, lags=35)  # Adjust the number of lags as needed
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.title('Autocorrelation Function (ACF)')
plt.show()

# Plot PACF
plot_pacf(ts, lags=35)  # Adjust the number of lags as needed
plt.xlabel('Lag')
plt.ylabel('Partial Autocorrelation')
plt.title('Partial Autocorrelation Function (PACF)')
plt.show()
```

```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt
from patsy import dmatrix
from sklearn.metrics import mean_squared_error
from pyearth import Earth
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.model_selection import train_test_split, cross_val_predict, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, mean_absolute_error, mean_squared_error, r2_score, \
roc_curve, auc, precision_score, recall_score, confusion_matrix
from sklearn.model_selection import KFold
from sklearn.tree import DecisionTreeRegressor,DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV, ParameterGrid, RandomizedSearchCV
from sklearn.ensemble import BaggingRegressor,BaggingClassifier,RandomForestRegressor,RandomForestClassifier
from sklearn.ensemble import VotingRegressor, StackingRegressor

from sklearn.linear_model import Ridge, RidgeCV, Lasso, LassoCV
import xgboost as xgb
from xgboost import XGBRegressor, XGBClassifier
```

```python
# suppressing warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
# splitting data into train and test
y = data.covid_19_deaths
X = data.drop('covid_19_deaths', axis = 1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=45)
```

```python
data.dtypes
```

```python
object_columns = data.select_dtypes(include=['object']).columns

print(object_columns)
```

```python
X_train['location'] = X_train['location'].astype('category')
X_train['date.y'] = X_train['date.y'].astype('category')
```

```python
X_test['location'] = X_test['location'].astype('category')
X_test['date.y'] = X_test['date.y'].astype('category')
```

```python
xgb = XGBRegressor(random_state=45, enable_categorical=True, tree_method='hist')

# Coarse grid search 
param_grid = {'max_depth':[4,6,8],
              'n_estimators':[800, 1000, 1200], 
              'learning_rate':[0.1,0.25,0.5],
              'reg_lambda':[0,1,10],
              'gamma':[0,10],
              'subsample':[0.4,0.45,0.5]
             }

cv = KFold(n_splits=2, shuffle=True, random_state=1)
grid = RandomizedSearchCV(xgb, param_grid, n_jobs=-1,verbose=1,cv=cv, scoring='neg_mean_absolute_error', random_state = 45)
grid.fit(X_train, y_train)

print(grid.best_score_, grid.best_params_)
```

```python
# Print best hyperparameters
print("Best hyperparameters:", grid.best_params_)

# Compute RMSE and R-squared on test data using the best model from grid search
best_model = grid.best_estimator_
y_test_pred = best_model.predict(X_test)
test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))
test_r2 = r2_score(y_test, y_test_pred)

print("Test RMSE:", test_rmse)
print("Test R-squared:", test_r2)
```
