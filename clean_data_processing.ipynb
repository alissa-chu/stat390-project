```python
import pandas as pd
```

```python
data = pd.read_csv('merged_clean.csv')
```

```python
data.head()
```

```python
contain_nulls = data.columns[data.isna().any()].tolist()
len(contain_nulls)
```

```python
print(contain_nulls)
```

```python
dict_null_columns = {}
for x in contain_nulls:
    dict_null_columns[x] = data[x].isna().sum()
```

# processing

```python
# imputing the distributed columns with 0 before December 2020 -> no vaccines yet available
```

```python
data.sort_values(by=['date.x'], inplace=True)
```

```python
dates = data.loc[~pd.isna(data['distributed'])]
```

```python
dates.head()
```

```python
dates['date.x'].count()
```

```python
columns_l = ['distributed',
 'distributed_janssen',
 'distributed_moderna',
 'distributed_pfizer',
 'distributed_novavax',
 'distributed_unk_manuf',
 'dist_per_100k',
 'distributed_per_100k_5plus',
 'distributed_per_100k_12plus',
 'distributed_per_100k_18plus',
 'distributed_per_100k_65plus']
```

```python
data.loc[pd.isna(data['distributed']), columns_l] = 0
```

```python
# dropping unnecessary columns -> duplicate values
```

```python
columns_to_drop = ['state', 'month.x', 'month.y', 'day_of_week']
data = data.drop(columns=columns_to_drop)
```

```python
# dummy variables
```

```python
data = pd.get_dummies(data, columns=['region', 'season'])
```
